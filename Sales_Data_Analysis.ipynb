```python
# Transformasi Data
# Memastikan data yang memiliki item_price < 0 atau total_price < 0
cek = retail_table.loc[(retail_table['item_price'] < 0) | retail_table['total_price'] < 0]
cek
```

The code creates a DataFrame named `cek` by filtering rows from the `retail_table` DataFrame where either the 'item_price' or 'total_price' is less than 0. This allows for the identification and inspection of rows with negative values in either of these columns. The resulting `cek` DataFrame contains rows with negative prices in either 'item_price' or 'total_price'.

```python
# Jika tidak masuk akal datanya dapat dibuang
if cek.shape[0] != 0:
    retail_table = retail_table.loc[(retail_table['item_price'] > 0) & (retail_table['total_price'] > 0)]

# Cek apakah masih ada order_id yang bernilai undefined dan delete row tersebut
cek = retail_table.loc[retail_table['order_id'] == 'undefined']
cek
```

The code checks if the DataFrame `cek` has rows (determined by `cek.shape[0] != 0`) where either 'item_price' or 'total_price' is less than 0. If such rows exist, it updates the `retail_table` DataFrame by keeping only the rows where both 'item_price' and 'total_price' are greater than 0.

Then, it checks if there are still rows in the `retail_table` DataFrame where 'order_id' is 'undefined', creating a new DataFrame `cek` with these rows. This suggests that rows with 'order_id' equal to 'undefined' are inspected, and if any exist, they are deleted from the `retail_table` DataFrame. This is a further data cleaning step to handle negative prices and undefined order IDs.

```python
# Jika ada maka buang baris tersebut
if cek.shape[0] != 0:
    retail_table = retail_table.loc[retail_table['order_id'] != 'undefined']
```

The code checks if there are rows in the DataFrame `retail_table` where the 'order_id' is 'undefined'. If any such rows exist, it updates the DataFrame by removing those rows. The condition `if cek.shape[0] != 0` ensures that the removal is performed only if there are rows with 'order_id' equal to 'undefined'. This is a data cleaning step to eliminate rows with undefined order IDs.

```python
# Filter hanya 5 province terbesar di pulau Jawa
print('\nFILTER 5 PROVINCE TERBESAR DI PULAU JAWA\n')
java = ['DKI Jakarta','Jawa Barat','Jawa Tengah','Jawa Timur','Yogyakarta']
retail_table = retail_table.loc[retail_table['province'].isin(java)]

# Untuk memastikan kolom provinsi isinya sudah sama dengan java
print(retail_table['province'].unique())
```

The code filters the `retail_table` DataFrame to include only the data for the five largest provinces in Java: DKI Jakarta, Jawa Barat, Jawa Tengah, Jawa Timur, and Yogyakarta. The `isin()` method is used to filter rows where the 'province' column matches any of the specified provinces in the `java` list. The second part of the code prints the unique values in the 'province' column after filtering to ensure that the data now only contains the specified provinces. This filtering is useful when focusing on a subset of provinces for further analysis.

```python
#Kelompokkan sesuai dengan order_date dan province kemudian aggregasikan
groupby_city_province = retail_table.groupby(['order_date','province']).agg({
   'order_id': 'nunique',
   'customer_id': 'nunique',
   'product_id': 'nunique',
   'brand': 'nunique',
   'total_price': sum
})
```

This code is using the `groupby` method to group the `retail_table` DataFrame by both 'order_date' and 'province'. Then, it uses the `agg` method to perform multiple aggregation functions on different columns. Here's your code:

Explanation:

- `groupby(['order_date', 'province'])`: This groups the DataFrame by both 'order_date' and 'province'.

- `.agg({...})`: This applies aggregation functions to different columns specified in the dictionary.

- `'order_id': 'nunique'`: Counts the number of unique order IDs within each group.

- `'customer_id': 'nunique'`: Counts the number of unique customer IDs within each group.

- `'product_id': 'nunique'`: Counts the number of unique product IDs within each group.

- `'brand': 'nunique'`: Counts the number of unique brands within each group.

- `'total_price': sum`: Calculates the sum of total prices within each group.

The resulting `groupby_city_province` DataFrame will have a multi-level index consisting of 'order_date' and 'province', and the specified aggregated metrics as columns.

```python
# Ubah nama kolomnya menjadi 'order','customer','product','brand','GMV'
groupby_city_province.columns = ['order','customer','product','brand','GMV']
groupby_city_province.head(10)
```

The code renames the columns of the DataFrame `groupby_city_province` to 'order', 'customer', 'product', 'brand', and 'GMV'. The `.head(10)` method is then used to display the first 10 rows of the DataFrame with the updated column names. This renaming is helpful for providing clearer and more meaningful column names for further analysis and interpretation.

```python
# Unstack untuk mendapatkan order_date di bagian baris dan province di bagian column
unstack_city_province = groupby_city_province.unstack('province').fillna(0)
unstack_city_province.head()
```

The code uses the `unstack` method on the `groupby_city_province` DataFrame to pivot the data, placing 'order_date' in the rows and 'province' in the columns. The resulting DataFrame, named `unstack_city_province`, is then filled with 0 for any missing values (NaN). This operation transforms the data, making it more suitable for analysis or visualization where 'order_date' and 'province' serve as the indices and columns, respectively. The `head()` method is used to display the first few rows of the resulting DataFrame.

```python
# Slicing data untuk masing-masing measurement, misal: order
idx = pd.IndexSlice['order']
by_order = unstack_city_province.loc[:,idx]
print('\nby order (5 data teratas):\n', by_order.head())
```

The code slices the `unstack_city_province` DataFrame to extract data for a specific measurement, in this case, 'order'. It uses `pd.IndexSlice` with the label 'order' to perform the slicing operation. The resulting DataFrame `by_order` contains data for the 'order' measurement. The `print()` statement displays the first 5 rows of this sliced DataFrame for inspection.

This allows you to examine the data specifically related to the 'order' measurement, providing insights into its structure and values.

```python
# Lakukan resampling pada data tersebut untuk dilakukan perhitungan rata-rata bulanan
by_order_monthly_mean = by_order.resample('M').mean()
print('\nby_order_monthly_mean (5 data teratas):\n', by_order_monthly_mean.head())
```



This code uses the `resample` method with the parameter `'M'` to resample the data at a monthly frequency and then calculates the mean for each month.

The resulting `by_order_monthly_mean` DataFrame will have the calculated monthly mean values.



```python
import matplotlib.pyplot as plt

# Plot untuk hasil resampling
by_order_monthly_mean.plot(
   figsize = (8,5),
   title = 'Average Daily order Size in Month View for all Province'
)
plt.ylabel('avg order size')
plt.xlabel('month')
plt.show()
```

The code uses Matplotlib to plot the results of resampling (`by_order_monthly_mean`). It creates a line plot with the average daily order size on the y-axis and months on the x-axis. The `figsize` parameter sets the size of the figure, and `title`, `ylabel`, and `xlabel` provide labels for the plot. This visualization helps to observe trends and patterns in the average daily order size across months for all provinces.

```python
# Create figure canvas dan axes for 5 line plots
fig, axes = plt.subplots(5, 1, figsize=(8, 25))

# Slicing index
idx = pd.IndexSlice
for i, measurement in enumerate(groupby_city_province.columns):
    # Slicing data untuk masing-masing measurement
    by_measurement = unstack_city_province.loc[:,idx[measurement]]
    # Lakukan resampling pada data tersebut untuk dilakukan perhitungan rata-rata bulanan
    by_measurement_monthly_mean = by_measurement.resample('M').mean()
    # Plot by_measurement_monthly_mean
    by_measurement_monthly_mean.plot(
        title = 'Average Daily ' + measurement + ' Size in Month View for all Province',
        ax = axes[i]
    )
    axes[i].set_ylabel('avg ' + measurement + ' size')
    axes[i].set_xlabel('month')

# Adjust the layout and show the plot
plt.tight_layout()
plt.show()
```

The code creates a subplot with five vertically stacked line plots, each representing one of the measurements ('order', 'customer', 'product', 'brand', 'GMV') for all provinces. It uses a loop to iterate through the measurements, slicing the data, performing monthly resampling, and plotting the average daily size for each measurement. The resulting visualization provides a comparative view of how each measurement changes over months for all provinces. The `tight_layout()` function improves the layout of the subplots for better readability.
